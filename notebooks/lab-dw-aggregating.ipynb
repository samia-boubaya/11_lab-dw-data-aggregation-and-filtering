#### Load and View Dataset

```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
import seaborn as sns
plt.style.use('ggplot')
pd.set_option('display.max_columns', 200)

# Load online_shoppers_intention.csv into a df_main DataFrame
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv")

# Display df DataFrame SHAPE
print("RAW Dataset:")
print(f"Dataset Shape: [{df.shape[0]}] ROWS and [{df.shape[1]}] COLUMNS\n")
display(df.head())
```

1. Create a new DataFrame that only includes customers who:
   - have a **low total_claim_amount** (e.g., below $1,000),
   - have a response "Yes" to the last marketing campaign.

```python
# Filter the DataFrame that includes customers with total_claim_amount below $1,000 AND response "Yes"
df_filtered = df[(df['total_claim_amount'] < 1000) & (df['response'] == 'Yes')]

# Display the filtered DataFrame
display(df_filtered.head())
```

```python
print("average monthly premium auto:", df["monthly_premium_auto"].mean())
display(df["monthly_premium_auto"].describe().round(2))


print(70*"-")
print("average customer lifetime value by policy:")
print(70*"-")
display(df["customer_lifetime_value"].groupby(df["policy"]).mean().round(2))


print(70*"-")
print("average customer lifetime value by policy:")
print(70*"-")
display(df_filtered["customer_lifetime_value"].groupby(df_filtered["policy"]).mean().round(2))
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
# number of customer grouped by state and policy
df_customers_states_policy = (
    df.groupby(["state", "policy"]).size().reset_index(name="customer_count")
)
display(df_customers_states_policy)
# total number of customers who have policies above 500 in each state
print("total number of customers who have policies above 500 in each state:")
df_customer_count_by_state = df_customers_states_policy[df_customers_states_policy["customer_count"] > 500].groupby("state")["customer_count"].sum()
display(df_customer_count_by_state)
```

```python
df_states = (
    df
    .groupby("state")
    .size()
    .reset_index(name="customer_count")
)

display(df_states)
```

**🔎 Business Insight:**

- California has the largest customer base by a wide margin, representing the primary market.

- All listed states meet the filtering condition of more than 500 customers, meaning no states were excluded after applying the threshold.

**📊 Business Insight:**

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.


```python
# display customer_lifetime_value by education level and gender
print(70*"-")
print("customer_lifetime_value by education level and gender:")
# mean customer_lifetime_value by education level and gender
display(df.groupby(["education", "gender"])["customer_lifetime_value"].mean().round(2))
print(70*"-")
# maximum customer_lifetime_value by education level and gender
display(df.groupby(["education", "gender"])["customer_lifetime_value"].max().round(2))
print(70*"-")
# minimum customer_lifetime_value by education level and gender
display(df.groupby(["education", "gender"])["customer_lifetime_value"].min().round(2))
```

---

**Data Insight:**
**1. Higher Education Is Associated With Higher CLV**
- Customers with Master’s and Doctor degrees have the highest lifetime values.
    - Master: F (2417.78), M (2272.31)
    - Doctor: F (2395.57), M (2267.60)
- This suggests that more educated customers tend to generate more long-term revenue.

**2. Bachelor and College Levels Have Lower CLV:**
- These groups may represent mid-value segments compared to postgraduate customers.
    - Bachelor and College degree holders show the lowest CLV overall.
    - Bachelor: F (1904.00), M (1898.01)
    - College: F (1898.68), M (1918.12)

**3. Gender Differences Vary by Education Level**
- Females have higher CLV than males in most education categories:
    - Bachelor,
    - Doctor,
    - High School or Below,
    - Master.
- Exception: In the College category, males (1918.12) slightly exceed females (1898.68).

**4. High School or Below Shows an Interesting Pattern**
- Females (2144.92) have noticeably higher CLV than males (1940.98).
- This gap is larger than in most other education categories.

**5. Highest and Lowest Segments**
- Highest CLV overall: Female customers with a Master’s degree (2417.78).
- Lowest CLV overall: Male customers with a Bachelor’s degree (1898.01).

**Business Insight:**
- Postgraduate customers (Master’s and Doctorate) are the most valuable segment.
- Female customers tend to have slightly higher lifetime value across most education levels.
- Targeted retention and premium offerings may be especially effective for highly educated female customers.
